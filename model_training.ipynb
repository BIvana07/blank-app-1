```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import pickle
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from copy import deepcopy

df1 = deepcopy(df)
df1.head()
```

```python
X = df1.drop(['Id', 'Species'],axis=1)
y = df1['Species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    random_state=0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import tree

le = LabelEncoder()

y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
sns.heatmap(X_train.corr(),annot=True)
```

```python
dtc = DecisionTreeClassifier(criterion='entropy',random_state=0)
dtc.fit(X_train,y_train)
```

```python
dtc_score_train = dtc.score(X_test,y_test)
print('Accuracy: ', dtc_score_train)
```

```python
dtc_score_test = dtc.score(X_test,y_test)
print('Accuracy: ', dtc_score_test)
```

```python
predict_test = dtc.predict(X_test)
print(classification_report(y_test,predict_test))
```

```python
plt.figure(figsize=(10,10))
cm = confusion_matrix(y_test, predict_test)
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
plt.figure(figsize=(10,10))
tree.plot_tree(dtc,filled=True, feature_names=['SepalLengthCm',
                                               'SepalWidthCm',
                                               'PetalLengthCm',
                                               'PetalWidthCm'])
```
